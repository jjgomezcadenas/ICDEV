```python
import numpy as np
import matplotlib.pyplot as plt

from invisible_cities.core.fit_functions import profileX
from invisible_cities.reco.dst_functions import load_dst

from invisible_cities.icaro.hst_functions import hist2d
from invisible_cities.icaro.hst_functions import labels


%matplotlib inline

plt.rcParams["figure.figsize"] = 12, 8
plt.rcParams[  "font.size"]    = 15
```

```python
dst = load_dst("/Users/Gonzalo/github/NEXTdata/Kr/dst_4734_filtered.root.h5", group="DST", node="Events")
X = dst.X.values
Y = dst.Y.values
Z = dst.Z.values
```

```python
Xrange = -200, 200
Yrange = -200, 200
Zrange =   50, 550

Xnbins = 200
Ynbins = 200
Znbins =  50

Xbins = np.linspace(*Xrange, Xnbins)
Ybins = np.linspace(*Yrange, Ynbins)
Zbins = np.linspace(*Zrange, Znbins)
```

```python
hist2d(Z, X, (Zbins, Xbins))
z, x, sx = profileX(Z, X, Znbins, Zrange, std=True)
plt.errorbar(z, x, sx, np.diff(z)[0]/2, "kp")
labels("Drift time (µs)", "x (mm)", "X vs Z")

hist2d(Z, Y, (Zbins, Ybins))
z, y, sy = profileX(Z, Y, Znbins, Zrange, std=True)
plt.errorbar(z, y, sy, np.diff(z)[0]/2, "kp")
labels("Drift time (µs)", "y (mm)", "Y vs Z")
```

```python
plt.plot(z, sx, "kp", label="x")
plt.plot(z, sy, "gp", label="y")
plt.legend()
labels("Drift time (µs)", "σ (mm)", "RMS vs Z")
```
